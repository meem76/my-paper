```python
#using outliers
# Defining Interquartile Range
q1 = train_data["Price"].quantile(0.25)
q3 = train_data["Price"].quantile(0.75)
iqr = q3 - q1
up = q3 + 1.5 * iqr
low = q1 - 1.5 * iqr
```

```python
def outlier_thresholds(dataframe, col_name, q1=0.25, q3=0.75):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit

outlier_thresholds(train_data, "Price")
```

```python
object_cols = [col for col in train_data.columns if train_data[col].dtypes == 'object']
object_cols
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_enc = LabelEncoder()

def obj_to_categorical(train_data, columns):
    for col in columns:
        train_data[col] = lbl_enc.fit_transform(train_data[col])

    return train_data

train_data = obj_to_categorical(train_data, object_cols)
```
